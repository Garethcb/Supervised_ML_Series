# <center> Naive Bayes Classifier
___

1. Naive Bayes is a probabilistic machine learning algorithm based upon the Bayes Theorem and sees its application in a wide variety of classification tasks.


2. Classification tasks such as:
    - Spam filtering
    - document classifiers
    - sentiment prediction
    - etc

___
## <center>Why Is It Naive?

Naive Bayes is naive because it assumes **independence** of input features. This means that the input features that go into our model are independent of one another (do not influence one another).
___    
#### <center> Example:
In spam filtering, the two messages below are treated the same due to the independence of each feature:
    - "Dear Mike, you won".
    - "Mike won, Dear you". 
___

## <center> Understanding Conditional Probability
<center> Consider a school with 110 students. These 110 students either prefer 'Soccer' or 'Golf' and are seen as either 'Females' or 'Males'.
    
___
#### <center> Worked Example
$$\begin{array}{c|c}
    & \text{Female} & \text{Male} & \text{Total} \\
  \hline
   \text{Soccer} & 31 & 41 & 72\\ 
  \hline
   \text{Golf} & 9 & 29 & 38\\
      \hline
   \text{Total} & 40 & 70 & 110
 \end{array}$$
___   
#### <center>  What is the conditional probability that a certain student likes ‘Golf’ given that he is a ‘Man’?

___   
$$P(Golf|Male) = \frac{P(Golf \bigcap Male)}{P(Male)} = \frac{29}{70} = 0.41$$
___  
Additionally, we can consider the conditional probability that a certain student is Male given that they like Golf.
    
Therefore, the Bayes Rule that we use for Naive Bayes, can be derived from these two notations:
___    
$$P(A|B) = \frac{P(A \bigcap B)}{P(B)}$$

$$P(B|A) = \frac{P(A \bigcap B)}{P(A)}$$
___

___
# <center> Bayes Theorom
___
$$P(A|B) = \frac{P(B|A).P(A)}{P(B)}$$
    
$$P(y|X) = \frac{P(X|y).P(y)}{P(X)}$$
___  
    - P(A|B) is the posterior probability
    - P(B|A) is the likelihood of evidence: conditional porbability of X given Y is of a particular class
    - P(A) is the prior
    - P(B) is the Probability of evidence and it remains static for all classes
___
1. The Bayes Rule allows us to use $P(X|Y)$ from our training dataset to find $P(Y|X)$.
    
    
    - Replacing A and B with X (input features) and Y (target class)


2. For our predictions based upon our testing set, our X is known while our target class is unknown. We can compute the probability of Y given X has already happened.



___

       \hline

 \end{array}$$

___
# <center> Mathematical Walkthrough
$$P(y|X) = \frac{P(X|y).P(y)}{P(X)}$$

$$X = (x_1,x_2,x_3,...,x_n)$$
___
1. Variable y: is the class variable (played golf), which represents if it is suitable to play golf or not given the conditions. 


2. Variable X: represents our input features (Outlook and windy)
___
$$P(y|x_1,...,x_n) = \frac{P(x_1|y).P(x_2|y).P(x_3|y)...P(x_n|y)}{P(x_1).P(x_2).P(x_3)...P(x_n)}$$
___
We can now substitute the values from our training set into the equation. Notably, the denominator remains static. Therefore, the denominator can be removed and a proportionality can be introduced:

$$P(y|x_1,...,x_n) \propto P(y)\prod_{i=1}^{n} P(x_i|y)$$
___
#### <center> For multiclass:

$$y = argmax_y P(y)\prod_{i=1}^{n} P(x_i|y)$$
    
___



#### <center> Golfing Dataset
$$\begin{array}{c|c}
    \text{Outlook} & \text{Windy} & \text{Played Golf} \\
  \hline
   \text{Rainy} & False & N \\ 
  \hline
   \text{Rainy} & False & Y \\
      \hline
   \text{Sunny} & False & Y \\
\hline
   \text{Rainy} & True & Y \\ 
  \hline
   \text{Overcast} & True & Y \\
      \hline
   \text{Overcast} & False & Y \\
       \hline
   \text{Sunny} & True & N 
 \end{array}$$

## <center> Played Golf
___
**Prior**:

P(A) = $\frac{Y}{Y+N}$ = $\frac{5}{7}$
___
**Likelihood of evidence $P(y|x_1)$**:

P(Played|Rainy) = $\frac{2}{5}$

P(Played|Sunny) = $\frac{1}{5}$

P(Played|Overcast) = $\frac{2}{5}$
___   
**Likelihood of evidence $P(y|x_2)$**:

P(Played|Windy) = $\frac{2}{5}$

P(Played|Not_Windy) = $\frac{3}{5}$


___
## <center> Not Played Golf
___   
**Prior**:
    
P(A) = $\frac{N}{Y+N}$ =$\frac{2}{7}$
___    
**Likelihood of evidence $P(y|x_1)$**:

P(Not_Play|Rainy) = $\frac{2}{2}$

P(Not_Play|Sunny) = $\frac{1}{2}$

P(Not_Play|Overcast) = $\frac{0}{2}$
___    
**Likelihood of evidence $P(y|x_2)$**:

P(Not_Play|Windy) = $\frac{1}{2}$

P(Not_Play|Not_Windy) = $\frac{1}{2}$
___

## <center> Will golf be played given that it is Sunny and Windy?
___

$$P(y|x_1,...,x_n) \propto P(y)\prod_{i=1}^{n} P(x_i|y)$$
    
$$P(A).P(Played|Sunny).P(Played|NotWindy)$$

$$y = \frac{5}{7}.\frac{1}{5}.\frac{3}{5} = 0.0857$$
___    
    
$$P(y|x_1,...,x_n) \propto P(y)\prod_{i=1}^{n} P(x_i|y)$$
    
$$P(A).P(\text{NotPlay|Sunny}).P(\text{Not Play|Not Windy})$$

$$y = \frac{2}{7}.\frac{1}{2}.\frac{1}{1} = 0.0714$$
___    
    
<center> Since Play is greater than Not Play we can classify this observation as played.

___
# <center> Laplace Correction
___
$$P(A).P(NotPlay|Overcast).P(NotPlay|NotWindy)$$

$$y = \frac{2}{7}.0.\frac{1}{1} = 0$$
___    
1. The value of P(Not_Play | Overcast) was zero in the above example.
    
    
2. Therefore, when you have a model that includes this feature, the entire probability will become zero. To avoid this, we increase the count of the variable with zero to a small value.

# <center> Gaussian Naive Bayes
    
- If we assume that the X follows a particular distribution, then you can plug in the probability density function of that distribution to compute the probability of likelihoods.

    
- If you assume the X’s follow a Normal (aka Gaussian) Distribution, which is fairly common, we substitute the corresponding probability density of a Normal distribution and call it the Gaussian Naive Bayes. You need just the mean and variance of the X to compute this formula.
___   
$$P(x_i|y) = \frac{1}{\sqrt{2\pi\sigma_y^2}} exp(-\frac{(x_i-y)^2}{2\sigma_y^2})$$
___

# <center> Gaussian Naive Bayes Prediction
___
Make a prediction for a person who eats:
          - 9 grams of popcorn
          - 63 grams of candy

````python
Image('GaussianSolutionPopcandy.png')
````

___
## <center> Likes Inception
___
**Prior**:

P(A) = $\frac{1}{2}$
___
**Likelihood of evidence**:

L(Popcorn = 9|Likes) = 0.021586

L(Candy = 63|Likes) = 4.802e-30
  
___
#### <center> To avoid underflow (accuracy of caluclation with really small numbers) we log our likelihoods
$$Log(P(A).L(Popcorn = 9|Likes).L(Candy = 63|Likes))$$
    
$$log(0.5)+log(0.021586)+log(4.802e-30) = -72$$
___
    
## <center> Disikes Inception
___
**Prior**:

P(A) = $\frac{1}{2}$
___
**Likelihood of evidence**:

L(Popcorn = 9|Dislikes) = 0.0519

L(Candy = 63|Dislikes) = 0.0119
    

___
#### <center> To avoid underflow (accuracy of caluclation with really small numbers) we log our likelihoods
$$Log(P(A).L(Popcorn = 9|Dislikes).L(Candy = 63|Dislikes))$$
    
$$log(0.5)+log(0.021586)+log(0.0119) = -5.78$$
___
#### <center> Since the score for dislikes Inception is greater than Likes inception we will label this point as Dislikes Inception
